# Reading Data from a CSV 

```python
#reading csv file

Dallasdata = pd.read_csv('https://raw.githubusercontent.com/aaroncfo/sql/main/combined_data_set_2016_2019.csv')
Dallasdata.head()
```

```python
# data type

Dallasdata.dtypes
```

```python
#creating a column called year extracting just the year from the date column 


Dallasdata['year'] = pd.DatetimeIndex(Dallasdata['OCCURRED_D']).year
```

```python
Dallasdata.head()
```

```python
# filtering out the 2019 data  

Dallasdata2 = Dallasdata[Dallasdata['year'] != 2019]
```

```python
Dallasdata2.head()
```

```python
# checking if all the year 2019 data are filterd out 
Dallasdata2['year'].unique()
```

```python
# Dropping the year column 

Dallasdata2.drop('year', inplace=True, axis =1)
```

```python
# Reading all the columns in the data
columns = Dallasdata2.columns
columns
```

```python
# dropping the "Unnamed: 0" column because it doesnt exist in the data we got from the API, that way we wont have any problems when we try to join the two data

Dallasdata2.drop('Unnamed: 0', inplace=True, axis =1)
```

```python
# making all the columns of the data we got from API lower case
df.columns =[x.lower() for x in df.columns]
```

```python
# making all the columns of the data we got from CSV lower case
Dallasdata2.columns =[x.lower() for x in Dallasdata2.columns]
```

```python
# checking to see if its converted to lower case
columns = Dallasdata2.columns
columns
```

```python
# checking to see if its converted to lower case
columns2 = df.columns
columns2

```

```python
# creating a copy of the API data with the same column order of the CSV data 
df2 = df[columns]
df2.head()
```

# Concating API data and CSV data

```python
# Creating the final dataset by concating the API and CSV data

final= pd.concat([Dallasdata2,df2], ignore_index=True)
final.head()
```
